```python
## Importing libraries required for this assignment
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Load the House data into a dataframe
home_df = pd.read_csv('austinHousing.csv') 
home_df.head(5)
```

```python
## Printing number of rows and columns
home_df.shape
```

```python
## Printing the dtype for each of the column
home_df.dtypes
```

```python
## Rounding the column average school rating to 2 decimal places
home_df['avgSchoolRating'] = home_df['avgSchoolRating'].round(2)
```

##### Questions to explore

1. Does school rating have impact on house price?
2. Which story of houses sold the most and the average sold price for each story house?

#### Question 1: Does school rating have impact on house price?

##### 2. Create a histogram or bar graph from your data.

```python
# Histogram function average school rating (avgSchoolRating )
home_df.hist(column='avgSchoolRating', figsize = (16,10),edgecolor='black')
plt.xlabel('Average School Rating')
plt.ylabel('# of Homes')
plt.title('House School Rating Distribution')
```

```python
## Histogram function on latest sales prices (latestPrice)
home_df.hist(column='latestPrice', figsize = (16,10),edgecolor='black')
plt.xlabel('Home Sales Price')
plt.ylabel('# of homes')
plt.title('House Sales Price Distribution')
```

#### 3. Create a boxplot from your data.

```python
# boxplot of average school rating (scale of 0-10) with respect to house price
bplot = home_df.boxplot(by ='avgSchoolRating', column =['latestPrice'], grid = False, figsize = (16,10))
bplot.set_ylabel('Home Price')
```

```python
# Calculate the engagement survey median score for the dataset
home_df['avgSchoolRating'].median()
```

#### 4. Create a bivariate plot from your data.

```python
# Scatterplot of Average School Rating by latest Price
plt.figure(figsize=(16, 10))
plt.scatter(home_df.avgSchoolRating, home_df.latestPrice)
plt.title('Average School Rating vs Home Price')
plt.xlabel('Average School Rating')
plt.ylabel('Home Price')
```

#### 5.Create any additional visualizations that will help to answer the question(s) you want to answer.
#### Line Chart

```python
## Creating a sun dataframe with average school rating and latest Price
home_sch_df=home_df[['avgSchoolRating', 'latestPrice']]
```

```python
## Calcualte median price for each school rating and plot line chart
home_sch_df.groupby(['avgSchoolRating']).median().plot(kind='line', y='latestPrice', figsize=(12,10), title='Median House Price by School Rating')
```

##### Hex Plot

```python
## Hex plot for average school rating and latest price
home_df.plot.hexbin(x='avgSchoolRating', y='latestPrice', gridsize=15, figsize = (16,10))
```
